```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
from openpyxl import Workbook
import time
import random

page=1
item=0
list = []

url = 'https://www.tripadvisor.com.tw/Restaurants-g293912-Tainan.html'
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

res = requests.get(url, headers=headers)
html = res.text

soup = BeautifulSoup(html, 'html.parser')

titlelist = soup.find_all('a', class_="_15_ydu6b")
for title in titlelist:
        urlshop = "https://www.tripadvisor.com.tw" + title.get('href')
        reshop = requests.get(urlshop, headers=headers)
        time.sleep(random.randint(1,4) )
        htmlshop = reshop.text
        soupshop = BeautifulSoup(htmlshop, 'html.parser')
        userid0=str(soupshop).split('uid":"',1)[1]
        userid=userid0.split('"',1)[0]#評論翻頁用UID
        gasl0=urlshop.split('-d',1)[1]
        gasl=gasl0.split('-R',1)[0]#評論翻譯用GASL
        
        shoptitle=soupshop.find('h1',class_="_3a1XQ88S")#店名
        list.append(shoptitle)
        shopranknum=soupshop.find('a',class_="_10Iv7dOs")#店評價數
        list.append(shopranknum)
        if (soupshop.find('span',class_="r2Cf69qf"))is not None:#確認有無總評價
            ranking=soupshop.find('span',class_=re.compile("^ui_bubble_rating bubble"))
                          
            if (soupshop.find('div',class_="jT_QMHn2"))is not None:#確認有無食物評價以及數量
                feel=len(soupshop.find_all('div',class_="jT_QMHn2"))
                for ranking in soupshop.find_all('span',class_=re.compile("^ui_bubble_rating bubble"),limit=1+feel):
                    list.append(re.sub(r'\D', "",str(ranking)))
                if feel==4:
                    continue
                else:
                    for m in range(1,5-feel):
                        list.append("無")
            else:
                list.append(ranking)
                for m in range(1,5):
                    list.append("無")    
        else:
            for m in range(1,6):
                list.append("無")        
        shopfoodtype=soupshop.find('div',class_="_1XLfiSsv")#店菜系
        list.append(shopfoodtype)
        shopwhere=soupshop.find('span',class_="_2saB_OSe")
        list.append(shopwhere)#店地址
        comment=[]
        commnum=soupshop.find_all('p',class_="partial_entry")
        if len(commnum)>0:
            pagenum=soup.find_all('div',class_="pageNumbers")
            if len(pagenum)==1:
                numb=soupshop.find_all('a',class_="pageNum")
                a=[]
                for num in numb:
                    a.append(int(num.text))
                for i in range(2,max(a)+1):
                    if i==max(a):
                        pageurl="https://www.tripadvisor.com.tw/GARecord?gac=Reviews&gaa=click&gal=last&gav={}&gani=false&gass=Restaurant_Review&gasl={}&gapu={}&gams=0"
                        pageres=requests.get(pageurl.format(max(a),gasl,userid), headers=headers)
                        pagehtml=pageres.text
                        pagesoup= BeautifulSoup(pagehtml, 'html.parser')
                        
                        
                    
            else:
                comments=soupshop.find_all('p',class_="partial_entry")
                for com in comments:
                    comment.append(com.text)
        else:
            list.append("無")
        list.append(comment)
        
        
    

    
```
